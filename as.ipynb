```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv('data/test.csv')
```

```python
data['Label'] = data['Label'].apply(lambda x: 1 if x == 'APT' else 0)
```

```python
data.iloc[5, -1] = 1
data.iloc[7, -1] = 0
data.iloc[18, -1] = 1
# data.iloc[14, -1] = 0
```

```python
data
```

```python
label = data['Label'].copy()
data = data.drop(columns=['Label'])
```

```python
model = RandomForestClassifier()
model.fit(data.iloc[:, 8:], label)
preds = model.predict(data.iloc[:, 8:])
```

```python
data['pred'] = preds
data['label'] = label
data
```

```python
import pickle
pickle.dump(model, open('model/MLP.model', 'wb'))
```
